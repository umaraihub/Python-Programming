## 📌 1. What is File Handling?

>File handling means reading from and writing to files stored on your computer using Python. Instead of losing data when a program ends, you can save it permanently in a file.

## 📌 2. File Opening Modes

**Before reading or writing, you must open a file using open() function.**
<br>
**Syntax**
```python
file = open("filename.txt", "mode")

**All File Modes**

| Mode      | Description                                                                                                                                                                           |
| --------- | ------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **`r`**   | Opens a file for **reading only**. The file **must already exist**, otherwise Python raises an error (`FileNotFoundError`). The file pointer starts at the **beginning of the file**. |
| **`w`**   | Opens a file for **writing only**. If the file already exists, its **previous content is deleted (overwritten)**. If the file does not exist, Python **creates a new file**.          |
| **`a`**   | Opens a file for **appending data**. New data is added **at the end of the file** without removing existing content. If the file does not exist, Python **creates it automatically**. |
| **`x`**   | Opens a file for **exclusive creation**. It **creates a new file**, but if the file already exists, Python raises a **FileExistsError**.                                              |
| **`r+`**  | Opens a file for **both reading and writing**. The file **must exist**. The file pointer starts at the **beginning**, so writing may overwrite existing content.                      |
| **`w+`**  | Opens a file for **reading and writing**. If the file exists, its **content is erased**. If it does not exist, a **new file is created**.                                             |
| **`a+`**  | Opens a file for **reading and appending**. New data is **added at the end of the file**. If the file does not exist, Python **creates it**.                                          |
| **`rb`**  | Opens a file for **reading binary data**. Used for files like **images, audio, videos, and PDFs**.                                                                                    |
| **`wb`**  | Opens a file for **writing binary data**. Existing file content is **overwritten**, or a new file is created.                                                                         |
| **`ab`**  | Opens a file for **appending binary data**. Binary content is added **at the end of the file** without deleting existing data.                                                        |
| **`rb+`** | Opens a file for **reading and writing binary data**. The file **must exist**.                                                                                                        |
| **`wb+`** | Opens a file for **writing and reading binary data**. If the file exists, its **content is overwritten**.                                                                             |
| **`ab+`** | Opens a file for **reading and appending binary data**. Binary data is **added to the end of the file**, and the file is created if it does not exist.                                |


## 📌 3. Opening & Closing a File

**✅ Method 1 — Manual open and close**

````python
file = open("course.txt","r")     # open file
# print(file)
print(file.read())     # read content
file.close()      # ALWAYS close file manually
````

**✅ Method 2 — using with statement (RECOMMENDED)**

````python
with open("course.txt","r") as file: 
    print(file.read())
# Automatically closes file — no need to call close()
````

## 📌 4. Writing to a File

**✅ 1.write() — Write a Single String**

````python
with open("python.txt","w") as file: # w mode — creates new file or OVERWRITES existing file
    file.write("Welcome to the Python Programming Language for AI Developers.")
    file.write("Master the core syntax, data structures, and OOP principles of Python.")
print("File written successfully.")
````

**✅ 2.writelines() — Write Multiple Lines at Once**

````python
content = [
    "Automate data extraction and preprocessing for training pipelines.",
    "Implement version control for AI models using Git and GitHub.",
    "Understand the ethical implications and safety of AI deployment.",
    "Develop custom chatbots and agents using LangChain and OpenAI.",
    "Deploy models to the cloud using Docker and modern API frameworks.",
    "Transform from a coder into an engineer capable of building the future." 
]
with open("data_source.txt","w") as file:
    file.writelines(content) #add content in single line
print("File written successfully.")
````

````python
python_content = [
    "Utilize Pandas to clean, transform, and analyze real-world data.",
    "Visualize insights and patterns using Matplotlib and Seaborn.",
    "Explore the fundamentals of Machine Learning with Scikit-learn.",
    "Dive into Deep Learning architectures using TensorFlow and PyTorch.",
    "Architect robust backends for scalable and efficient AI models."
]
with open("learning.txt","w") as file:
    for i in python_content:
        file.writelines(i + "\n") #add content in separate line by concatenating \n
print("File written successfully.")
````

## 📌 5. Reading from a File

**✅ 1.read() — Read Entire File at Once**

````python
with open("course.txt","r") as file:
    data = file.read()
    print(data)
````

````python
with open("learning.txt","r") as file:
    data = file.read()   # Read entire file content as one string
    print(data)
````

**✅ 2.read(n) — Read Only N Characters**

````python
with open("data_source.txt","r") as file:
    characters = file.read(10)  # read first 10 characters only
    print(data)
````

**✅ 3.readline() — Read One Line at a Time**

````python
with open("learning.txt","r") as file:
    line1 = file.readline()   # reads first line
    line2 = file.readline()   # reads second line
    print(line1)
    print(line2)
````

**✅ 4.readlines() — Read All Lines as a List**

````python
with open("learning.txt","r") as file:
    data = file.readlines()    # returns list of all lines
    print(data)
````

````python
with open("learning.txt","r") as file:
    data = file.readlines()
    # Loop through each line
    for line in data: #get first line of list then 2 elements or line of the list and so on
        # print(d)
        print(line.strip())  # strip() removes \n at end
````

**Loop Through File Directly (Best Way)**

````python
with open("learning.txt","r") as file:
    for line in file:
        print(line.strip())
````

## 📌 6. Appending to a File

````python
# a mode — adds to END of file, does NOT overwrite
with open("data.txt","a") as file:
    file.write("Favorite Subject: Python")

# Now read to verify
with open("data.txt","r") as file:
    print(file.read())
````

## 📌 8. File Pointer — seek() and tell()

````python
with open("newfile.txt","r") as file:
    print(file.tell())   # 0 — at beginning
    
    content = file.read(6)  # read 6 characters
    print(content)
    
    print(file.tell())   # 6 — moved 6 positions
    file.seek(0)
    print(file.tell())
    file.seek(10)
    print(file.tell())
    print(file.read())  #Reading starts from position 10 until the end.
````
