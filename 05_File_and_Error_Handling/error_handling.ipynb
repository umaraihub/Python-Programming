## 📌 10. Error Handling with Files

**✅Always handle file errors using try/except**
<br>
**File error handling is used to prevent a program from crashing when file-related errors occur.**
<br>
**Syntax:**
```python
try:
    # code that may cause an error
except ErrorType:
    # code to handle the error

````python
try:
    print("Hello World")
    print(x) #x is not defined so exception occur in try block and accept except block 
except Exception:  # it catches almost all common errors. Instead of catching all errors, it is usually better to catch specific errors.
    print("An Exception Occurred")
````

````python
try:
    print(x)
except Exception as e:
    print("Error:", e)  #e stores the actual error message.
````

````python
# print(x) # NameError --> crash the program while i print x variable
````

````python
try:
    print(x)
except NameError:
    print("Variable x is not defined")
````

````python
try:
    print("Hello World")
except NameError:
    print("Variable name is not defined")
except ValueError:
    print("Enter Valid Value")
except FileNotFoundError:
    print("File Does not Exist")
except ZeroDivisionError:
    print("Not divisible by zero")
else:   #else block executes only if no error occurs in the try block.
    print("No error occurs in the try block")
finally:
    print("block always executes regardless of whether an error occurs or not.")
````

````python
try:
    with open("data.txt","r") as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    print("File Not Found")
````

````python
mode = input("Enter file mode (r/w/a/x): ")

try:
    with open("course.txt", mode) as file:

        if mode == "r":
            data = file.read()
            print("\nFile Content:")
            print(data)

        elif mode == "w":
            text = input("Enter text to write in file: ")
            file.write(text)
            print("Data written successfully.")

        elif mode == "a":
            text = input("Enter text to append in file:")
            file.write(text)
            print("Data appended successfully.")

        elif mode == "x":
            print("New file created successfully.")

except PermissionError:
    print("No permission to access file.")

except IOError:
    print("Input/Output error during file operation.")

except Exception as e:
    print("Unexpected Error:", e)

else:
    print("File operation completed successfully.")

finally:
    print("Program execution finished.")
````
