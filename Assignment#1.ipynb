## 5. Write a Python program which accepts the user's first and last name and print them in reverse order with a space between them.

## 7. Write a program which takes 5 inputs from user for different subjectâ€™s marks, total it and generate mark sheet using grades?

```python
print("WELCOME TO UMAR'S GRADING SYSTEM")

print("Enter Marks of 5 Subjects: (1.English, 2.Mathematics, 3.Physics, 4.Chemistry, 5.Computer)")
Eng = float(input("Enter English Marks = "))
Maths = float(input("Enter Mathematics Marks = "))
Phy = float(input("Enter Physics Marks = "))
Chem = float(input("Enter Chemistry Marks = "))
Comp = float(input("Enter Computer Marks = "))

Obt_Marks = Eng + Maths + Phy + Chem + Comp
print("Obtained Marks = ", Obt_Marks)

percentage = (Obt_Marks/500)*100
print("Percentage = ", round(percentage,2))

if(percentage >= 90 and percentage <= 100):
    print("Grade A+")
    
elif(percentage >= 80 and percentage <=89):
    print("Grade A")
    
elif(percentage >= 70 and percentage <=79):
    print("Grade B")
    
elif(percentage >= 60 and percentage <=69):
    print("Grade C")
    
elif(percentage >= 50 and percentage <=59):
    print("Grade D")
    
elif(percentage >= 40 and percentage <=49):
    print("Grade E")
    
elif(percentage < 40):
    print("Grade F")
    
else:
    print("Invalid Marks! Please enter marks between 0 and 100.")
```

## 8. Write a program which take input from user and identify that the given number is even or odd.

```python
# Number is Even or Odd

num = int(input("Enter a Number = "))

if(num % 2 == 0):
    print("Number is Even")
else:
    print("Number is Odd")
```
