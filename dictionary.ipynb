## Dictionary Notes

```python
#A dictionary is a collection of key-value pairs.
#Keys are unique.
#Values can be anything: numbers, strings, lists, other dictionaries, etc.

student = {
    "Name": "Umar",
    "Age" : 26,
    "Course" : ["Python","C/C++","JavaScript"]
}
print(student)
print(type(student)) #data type of dictionary
```

## Accessing Values

```python
#Accessing Values
print(student["Name"]) #using key

print(student.get("Age")) #Get the value of the Age key
```

```python
# Access Values
values = student.values()
print(values)
```

```python
#Access Keys
keys = student.keys()
print(keys)
```

```python
#Access Keys and Values
keys_values = student.items()
print(keys_values)
```

```python
#Access Keys Using for loop 
for i in student:
    print(i)
```

```python
# Access Values using for loop
for i in student:
    print(student[i])
```

```python
#Access keys and values with Pair
for key,value in student.items():
    print(key,":",value)
```

## Adding Items

```python
# Adding Items
student["Class"] = 12 #Add a new key-value
print(student)

student.update({"Id":21}) #using update method --> add key (Id) and value(21)
print(student)
```

## Update Exisiting Values

```python
#Update existing key
student["Age"] = 25
print(student)
```

```python
student.update({"Class":11})
print(student)
```

## Removing Items

```python
# Removing Items
student.pop("Class") #Remove by key
print(student)
```

```python
student.popitem() #removes the last inserted item 
print(student)
```

```python
del student["Age"]
print(student)
```

## Calculate Items in Dictionary

```python
print(student)
print(len(student)) #how many items a dictionary
```

## Dictionary inside the List (Quiz Application)

```python
# Quiz Application Using Dictionary and List

# List of questions (each question is a dictionary)
quiz_questions = [
    {
        "question": "1. What is the correct file extension of Python?",
        "options": ["A) .pt", "B) .py", "C) .python", "D) .p"],
        "answer": "B"
    },
    {
        "question": "2. Which data type stores multiple values?",
        "options": ["A) int", "B) float", "C) list", "D) bool"],
        "answer": "C"
    },
    {
        "question": "3. Which symbol is used for comments in Python?",
        "options": ["A) //", "B) #", "C) <!--", "D) **"],
        "answer": "B"
    },
    {
        "question": "4. What does len() function do?",
        "options": ["A) Add values", "B) Delete value", "C) Count length", "D) Sort list"],
        "answer": "C"
    },
    {
        "question": "5. Which keyword is used for loop?",
        "options": ["A) loop", "B) repeat", "C) for", "D) iterate"],
        "answer": "C"
    }
]

score = 0

print("📘 Welcome to Python Quiz App\n")

# Loop through each question
for q in quiz_questions:
    print(q["question"])
    
    for option in q["options"]:
        print(option)
    
    user_answer = input("Enter your answer (A/B/C/D): ").upper()
    
    if user_answer == q["answer"]:
        print("✅ Correct!\n")
        score += 1
    else:
        print("❌ Wrong! Correct answer is:", q["answer"], "\n")

# Final Result
print("🎯 Quiz Completed!")
print("Your Final Score:", score, "/", len(quiz_questions))

percentage = (score / len(quiz_questions)) * 100
print("Percentage:", percentage, "%")

if percentage >= 70:
    print("Result: Pass 🎉")
else:
    print("Result: Fail ❌")
```
