## Print series and asterisk pattern using for loop

```python
#A for loop iterates over a sequence (like a list, string, tuple, or range) and executes a block of code for each item.
#for item in sequence:
    # set of statements

for i in range(1,5):
    print(i)

for i in range(1,5): #start=1 stop = 5(exclude last value 5) increment 1 by default
    print(i*"*")
```

```python
for i in range(5,0,-1): # Loop from 5 down to 1 (greater than 0 or exclude 0), decreasing by 1 each time
    print(i)

for i in range(5,0,-1):
    print(i*"*")
```

```python
n = int(input("Enter Number of Rows = "))  # number of rows in the top half

# Loop through each row
for i in range(1, 2*n):
    if i <= n:
        # Top half
        for j in range(n - i):
            print(" ", end="")  # print spaces
        for j in range(i):
            print("* ", end="")  # print stars with space
    else:
        # Bottom half
        for j in range(i - n):
            print(" ", end="")  # print spaces
        for j in range(2*n - i):
            print("* ", end="")  # print stars with space
    print()  # new line (print() at the end of outer loop)
```

```python
n = int(input("Enter number of rows: "))

for i in range(n, 0, -1):
    # Print spaces
    for j in range(n - i):
        print(" ", end="")
    # Print stars
    for j in range(i):
        print("*", end="")
    print()  # move to next line
```

```python
#Second Method
n = int(input("Enter number of rows: "))
for i in range(n, 0, -1):
    print(" " * (n - i) + "*" * i)
```

```python
#Write a Python program that takes a list of numbers and prints whether each number is even or odd.
numbers = [1, 4, 7, 10, 13]
for num in numbers:
    if (num % 2 == 0):
        print(num, "is even")
    else:
        print(num, "is odd")
```

```python
#Write a Python program that takes a list of numbers and prints whether each number is positive, negative, or zero.
numbers = [1, 4, -7, -10, 0]
for num in numbers:
    if (num > 0):
        print(num, "is Positive")
    elif (num < 0):
        print(num, "is Neagtive")
    else:
        print(num, "is Zero")
```

```python
#Write a Python program that takes a list of strings and prints whether each string contains the letter "a" or not.

# List of fruits
fruits = ["Apple", "Kiwi", "Cherry", "Orange"]

for value in fruits:
    if ("a" in value.lower()):  # convert to lowercase before checking
        print(value, "contains the letter 'a'")
    else:
        print(value, "does not contain the letter 'a'")
```

## While Loop

```python
#A while loop repeats a block of code as long as a condition is True.
#Syntax
#while condition:
    # set of statements

i = 1 #Initialization
while (i < 5): #Test Condition
    print(i) #statements
    i += 1  # increment i to avoid infinite loop (increment/decrement)

#asterisk pattern
i = 1
while (i < 5): #condition is true when value is less than 5
    print(i*"*")
    i += 1
```

```python
i = 4 #Initialization
while (i > 0): #Test Condition
    print(i) #statements
    i -= 1  # decrement i to avoid infinite loop (increment/decrement)

#asterisk pattern
i = 4
while (i > 0): #condition is true when value is greater than 5
    print(i*"*")
    i -= 1
```

```python
user_input = input("Enter any input value = ")

while user_input.lower() != "exit":
    user_input = input("Type 'exit' to stop: ")
    print("You typed:", user_input)
```

```python
# Program to check grades using while loop
marks = int(input("Enter your marks (0-100): "))

while marks >= 0:  # loop until user enters a negative number
    if marks >= 90:
        print("Grade: A+")
    elif marks >= 80:
        print("Grade: A")
    elif marks >= 70:
        print("Grade: B")
    elif marks >= 60:
        print("Grade: C")
    elif marks >= 50:
        print("Grade: D")
    else:
        print("Grade: F")
    
    # Ask for next input
    marks = int(input("Enter your marks (negative to exit): "))
```

## Nested Loop

```python
#A nested loop is a loop inside another loop. The inner loop runs completely every time the outer loop runs once.
for i in range(1, 4):      # Outer loop --> runs 3 times (i = 1, 2, 3)
    for j in range(1, 4):  # Inner loop --> runs 3 times for each i
        print(f"{i} x {j} = {i*j}") 
        #f-string (formatted string) allows you to embed variables directly inside a string using curly braces {}
    print("---------")  # Separator for each i
```

```python
first_name = ["Areej", "Muhammad", "Maryam", "Ali"] # List of first names
last_name = ["Fatima", "Umar", "Malik", "Raza"] # List of last names
full_name = [] # Empty list to store full names
for first in first_name: # Outer loop: take each first name from the first_name list
    for last in last_name:   # Inner loop: take each last name from the last_name list
        full_name.append(first + " " + last)  # Combine first and last name and add to the full_name list
print(full_name) # Print the complete list of full names
```
