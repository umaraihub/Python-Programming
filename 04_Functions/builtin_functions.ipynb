## 📌 6. Built-in Functions

**Python has many ready-made functions available without importing anything.**

## enumerate() — Index + Value Together

```python
fruits = ["Apple", "Banana", "Mango"]
for index, fruit in enumerate(fruits): # default starts from 0
    print(index,":", fruit)
# 0 Apple
# 1 Banana
# 2 Mango
print()
# start index from 1
for index, fruit in enumerate(fruits, start=1):
    print(f"{index} : {fruit}")
# 1. Apple
# 2. Banana
# 3. Mango
```

## zip() — Combine Two or More Lists

```python
names  = ["Ali",  "Sara",  "Ahmed"]
marks  = [85,     92,      78]
grades = ["A",    "A+",    "B"]

for name, mark, grade in zip(names, marks, grades):
    print(f"{name} : {mark} : {grade}")
# Ali   : 85 : A
# Sara  : 92 : A+
# Ahmed : 78 : B
```

## id() and help()

```python
x = 10
print(id(x))       # memory address of variable e.g. 140732

help(print)        # full documentation of print function
help(len)          # full documentation of len function
```
