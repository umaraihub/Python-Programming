## 📌 1. Functions — def

**A function is a reusable block of code that performs a specific task. You write it once and call it as many times as needed anywhere in your program.**

**Syntax:**
```python
def function_name():
    # block of code
```
**This statement defines a function. The function executes only when it is called.**

````python
def message(): # Define function
    print("Welcome to Umar's AI Hub")
message() # Call function
````

````python
def student():
    name  = "Arslan"
    age   = 27
    print("Name : ",name,"\nAge : ",age) 
student()
````

## 📌 2. Function Parameters & Arguments

**Parameter** - A parameter is the variable listed inside the parentheses in the function definition.
<br>
**Argument** - An argument is the actual value that is sent to the function when it is called.

**1. Positional Arguments — order matters**

````python
def student(name,age): #name and age (parameters)
    print("Name = ", name ,"\nAge =", age)
student("Arslan","27") #Arslan and 27 (Argument)
````

**2. Keyword Arguments — order does NOT matter**

````python
def student(name,age): 
    print("Name = ", name ,"\nAge =", age)
student(name="Asma",age=25) 
student(age=22,name="raza") 
````

**3. Default Arguments — value used if not provided**

````python
def student(name,age=25):
    print("Name : ",name,"\nAge : ",age) 
student("Arslan",27)
student("Zarnish")
````

**4. Arbitrary Arguments \*args — Accept Multiple Positional Arguments**

````python
# When you don't know how many arguments will be passed
# *args stores all values as a TUPLE
def student(name,*skills):
    # print("Name : ",name,"\nSkills : ",skills)
    print("Name : ", name)
    print("Skills :")
    for skill in skills:
        print("-",skill)
student("Arslan","Python","Gen AI","Prompt Engineering")
````

**5. Arbitrary Keyword Arguments `**kwargs` — Accept Multiple Keyword Arguments**

````python
# kwargs stores all values as a DICTIONARY
def student(**info):
    print(info)
    for key,value in info.items():
        print(key,":",value)
student(name = "Arslan",age = 27, city = "Islamabad")
````

## 📌 3. Return Statement

**A return statement is used inside a function to send a value back to the place where the function was called.**
<br>
**Syntax:**
```python
def function_name():
    return value

````python
def add(a,b):
    add = a + b
    return add  # sends the sum back to where the function is called
num1 = int(input("Enter First Number = "))
num2 = int(input("Enter Second Number = "))
result = add(num1,num2)  # calls the function and stores returned value
print("SUM =", result)
````
