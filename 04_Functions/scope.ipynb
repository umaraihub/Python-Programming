## 📌 4. Scope — Local & Global Variables

**Local Variable**

```python
def check():
    x = 10          # local variable
    print(x)        # works fine — inside function
check()             # 10
# print(x)          # NameError — x not accessible outside
```

**Global Variable**

```python
name = "Muhammad Umar"        # global variable
def show():
    print(name)     # accessible inside function
show()              # Muhammad Umar
print(name)         # Muhammad Umar
```
