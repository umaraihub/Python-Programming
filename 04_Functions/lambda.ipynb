## 📌 5. Lambda Functions

**A lambda function is a small anonymous function. 
A lambda function can take any number of arguments, but can only have one expression.**

**Syntax**
```python
   lambda arguments : expression

````python
add = lambda a, b : a + b

print("SUM :",add(5, 3))
````

````python
# Square of a number
square = lambda x : x ** 2
n = int(input("Enter a Number : "))
print(f"Square of {n} = {square(n)}")
````

**Lambda with map() — Apply Function to Every Item**

````python
# The map() function applies a function to every item in an iterable:
numbers = [1,2,3,4,5]
print("Original List : ",numbers)
squares = list(map(lambda n : n**2, numbers)) # squares of all numbers
print("Square : ",squares)    # [1, 4, 9, 16, 25]
````

**Lambda with filter() — Filter Items by Condition**

````python
numbers = []
for i in range(10):
    num = int(input(f"Enter Number {i+1}:"))
    numbers.append(num)
print("Final Numbers List = ",numbers)
# filter(function, list) — keeps only items where function returns True
result = list(filter(lambda x : x%2 == 0,numbers))
print("Even Numbers List  :",result)
````

**Lambda with sorted() — Custom Sorting**

**The sorted() function sorts any iterable (like a list or tuple) and can take a key function to determine the sort order. A lambda function is often used as the key argument to customize sorting.**
<br>
**Syntax:**
```python
sorted(iterable, key=lambda variable_name: expression, reverse=False)

````python
print(numbers)
````

````python
result = sorted(numbers,key=lambda x:x*x,reverse = False) #Sorting is done based on x*x, so smaller squares come first.
print(result)
````
