## 1. Make a calculator using Python with addition, subtraction, multiplication, division and power.(Remainder and Floor Division)

```python
print("WELCOME TO UMAR'S SIMPLE CALCULATOR")
```

```python
print("Available Operators: +  -  *  /  %  //  **")
num1 = float(input("Enter First Number: "))
num2 = float(input("Enter Second Number: "))
operator = input("Enter Operator (+, -, *, /, %, //, **): ")
print("Num1 = ",num1,"\nNum2 = ",num2,"\nOperator = ",operator )

if operator == '+':
    result = num1 + num2
    print("Sum =", result)

elif operator == '-':
    result = num1 - num2
    print("Difference =", result)

elif operator == '*':
    result = num1 * num2
    print("Product =", result)

elif operator == '**':
    result = num1 ** num2
    print("Exponentiation =", result)

elif operator == '/':
    if num2 != 0:
        result = num1 / num2
        print("Division =", result)
    else:
        print("Error (ZeroDivisionError)! Division by zero is not allowed.")

elif operator == '%':
    if num2 != 0:
        result = num1 % num2
        print("Remainder =", result)
    else:
        print("Error (ZeroDivisionError)! Modulus by zero is not allowed.")

elif operator == '//':
    if num2 != 0:
        result = num1 // num2
        print("Floor Division =", result)
    else:
        print("Error (ZeroDivisionError)! Floor division by zero is not allowed.")

else:
    print("Invalid Operator! Please use only +, -, *, /, %, //, **")
```

## 2. Write a program to check if there is any numeric value in list using for loop.

```python
my_list = ["Ali",5.4,2,'c']
for i in my_list:
    if type(i) == int or type(i) == float:
        print("List Contains a Numeric Value")
        break;
else:
    print("The list does not contain any numeric value.")
```

## 2. 2nd Method

```python
my_list = ["Ali", 5.4, 2, 'c']

for i in my_list:
    if isinstance(i, (int, float)):  #isinstance(object, datatype). It checks whether a variable belongs to a specific data type.
        print("List Contains a Numeric Value")
        break
else:
    print("The list does not contain any numeric value.")
```

## 5. Write a program to identify duplicate values from list.

```python
my_list = [10, 20, 3, 2, 30, 20, 1,3]
duplicates = []
for item in my_list:
    if my_list.count(item) > 1 and item not in duplicates: 
        #Check if item appears more than once and is not already added to duplicates list
        duplicates.append(item)
print("Duplicate values are:", duplicates)
```
