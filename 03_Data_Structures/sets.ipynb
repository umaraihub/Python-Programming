## 📌 3. Set

```python
# A set is an unordered collection of unique items used to store multiple values
# in a single variable.

# Sets do not allow duplicate values and items are not stored in a fixed order.

# Syntax:
# set_name = {item1, item2, item3, ...}


# Creating sets
fruits = {"Apple", "Banana", "Mango", "Orange"}   # Set of strings
marks  = {50, 20, 40, 40, 80}                        # Set of integers. Sets automatically remove duplicate values
mixed = {"Ali", 25, 3.5, True}                    # Different data types
empty_set = set()                           # Creating Empty Set. Only Curly Braces {} creates an empty dictionary, not a set.
 
print(fruits)
print(marks)
print(mixed) 
print(empty_set) 
```

```python
# Accessing Set Elements
fruits = {"Apple", "Banana", "Mango"}
# Sets are unordered, so they cannot be accessed using index.
for fruit in fruits:
    print(fruit)
```

## Set Functions (Built-in Functions)

## 1. add()

```python
fruits = {"Apple", "Banana"}
fruits.add("Mango") # Add a single element to the set
print(fruits)
```

## 2. update()

```python
fruits = {"Apple", "Banana"}
fruits.update(["Mango", "Orange"]) # Add multiple elements to the set using update()
print(fruits)
```

## 3. remove()

```python
marks  = {50, 20, 40, 40, 80}
marks.remove(20) # Remove element 20 from the set. Error occurs if the element does not exist.
print(marks)
```

## 4. discard()

```python
marks  = {50, 20, 40, 40, 80}
marks.discard(50)   # discard() removes the element if present. No error even if element is not present
print(marks)
```

## 5. pop()

```python
marks  = {50, 20, 40, 40, 80}
marks.pop() #Removes a random element because sets are unordered.
print(marks)
```

## 6. len()

```python
marks  = {50, 20, 40, 40, 80}
print(len(marks)) #len() returns the number of elements in the set
```

## 7. clear()

```python
marks  = {50, 20, 40, 40, 80}
marks.clear() # clear() removes all elements from the set
print(marks)
```

## Set Operations

## 1. union()

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1.union(set2)) #Combines elements of two sets.
```

## 2. intersection()

```python
set1 = {1, 2, 3}
set2 = {2, 3, 4}
print(set1.intersection(set2)) #Common elements between sets.
```

## 3. difference()

```python
set1 = {1, 2, 3}
set2 = {2, 3, 4}
print(set1.difference(set2)) #Elements present in first set but not in second.
```
