## 📌 Dictionary in Python

**"A dictionary is an unordered, mutable collection of data that stores values in key–value pairs."**
**1.Each value is accessed using a unique key.**
**2.Keys must be unique and immutable (string, number, tuple).**
**3.Values can be any data type.**

**Syntax:**

```python
dictionary_name = { "key1": value1, "key2": value2, "key3": value3 }

**Creating a Dictionary**

````python
# String Values 
student = {
    "Name"  : "Muhammad Umar",
    "City"  : "Islamabad",
    "Grade" : "A"
}

#Integer Values
marks = {
    "Math"     : 95,
    "Physics"  : 88,
    "Computer" : 92
}

#Mixed Values
profile = {
    "name"       : "Ali Hassan",   # string
    "age"        : 17,             # integer
    "gpa"        : 3.8,            # float
    "is_present" : True            # boolean
}

#Empty Dictionary
car = {}

#Single Item
country = {
    "name" : "Pakistan"
}
````

````python
student = {
    "Name": "Umar",
    "Age" : 26,
    "Subjects": ["Math", "Physics", "Computer"],   # List inside dictionary
    "Marks": (85, 90, 88)                          # Tuple inside dictionary
}
print(student)
print(type(student))
````

**Accessing Values**

````python
print("==========Method 1 -- key_name==========")
print(student["Name"])       # Muhammad Umar
print(student["Subjects"])   # ['Math', 'Physics', 'Computer']
print(student["Subjects"][1]) # Physics

print("==========Method 2 -- get()==========")

print(student.get("Age"))    # 26
print(student.get("Subjects"))    # ["Math", "Physics", "Computer"]
print(student.get("Subjects")[0])    # Math
print(student.get("City"))   # None — no error if key missing
````

**Looping through dictionary**

````python
student = {
    "Name": "Umar",
    "Age" : 26,
    "Subjects": ["Math", "Physics", "Computer"],   # List inside dictionary
    "Marks": (85, 90, 88)                          # Tuple inside dictionary
}
print(student)
print() #line Break or new line
print("Keys only")
for i in student: # keys only
    print(i)
print()
print("Values only")
for i in student: # Values only
    print(student[i])
````

## 📌 Methods / Functions

## 1. keys()

````python
student = {
    "Name": "Umar",
    "Age" : 26,
    "Subjects": ["Math", "Physics", "Computer"],   # List inside dictionary
    "Marks": (85, 90, 88)                          # Tuple inside dictionary
}

print(student.keys())          
print(list(student.keys()))

for key in student.keys():
    print(key)
````

## 2. values()

````python
print(student.values())
print(list(student.values()))

for value in student.values():
    print(value)
````

## 3. items()

````python
print(student.items())
print(list(student.items()))

for key,value in student.items():
    print(key, ":", value)
````

## 4. get()

````python
student = {
    "Name": "Umar",
    "Age" : 26,
    "Subjects": ["Math", "Physics", "Computer"],   # List inside dictionary
    "Marks": (85, 90, 88)                          # Tuple inside dictionary
}

print(student.get("Name"))
print(student.get("Subjects"))
print(student.get("Marks"))
print(student.get("City"))
print(student.get("Address","Not Found"))
````

## 5. update()

````python
student.update({"Name":"Muhammad Umar"})

marks = list(student["Marks"])
marks[1] = 95
student["Marks"] = tuple(marks)

student.update({"City":"Islamabad","Address":"Near Rawalpindi"})

print(student)
````

## 6. pop()

````python
removed = student.pop("Address")
print(removed)
print(student)
````

## 7. popitem()

````python
last_item = student.popitem()
print(last_item)
print(student)
````

## 8. len()

````python
print(len(student))
````

## 9. copy()

````python
student_copy = student.copy()
student_copy["Name"] = "Areej"
print("Original:", student)       # still MUhammad Umar
print("Copy    :", student_copy)  # Areej
````

## 10. clear()

````python
student_copy.clear()
print(student_copy)
````

## 📌 Adding, Updating & Deleting

````python
student = {
    "Name": "Umar",
    "Age" : 26,
    "Subjects": ["Math", "Physics", "Computer"],   # List inside dictionary
    "Marks": (85, 90, 88)                          # Tuple inside dictionary
}

# Add a new key-value pair
student["City"] = "Islamabad"
print(student)   # city is added

# Update an existing key's value
student["Age"] = 27
print(student)   # age is now 21

# Delete a specific key using del
del student["City"]
print(student)   # city is removed

# Delete entire dictionary from memory
del student
print(student)
# print(student)  # NameError: student is not defined
````

## 📌 Nested Dictionary

````python
school = {
    "student1" : {
        "name" : "Ali",
        "age" : 24,
        "subjects" :["Computer","Physics","Chemistry"]
    },
    "student2" : {
        "name" : "Asma",
        "age" : 23,
        "subjects" :["Computer","Physics","Chemistry"]
    },
    "student3" : {
        "name" : "Asad",
        "age" : 21,
        "subjects" :["Biology","Physics","Chemistry"]
    }
}
print(school)
print()
print(school.items())
# print()
# print(list(school.items()))
print()

for student in school:
    print(student)
    print(school[student])

print()

for students,info in school.items():
    print(students,":",info)
````

## 📌 QUIZ APPLICATIONS (Project)

````python
# Quiz Application Using Dictionary and List

# List of questions (each question is a dictionary)
quiz_questions = [
    {
        "question": "1. What is the correct file extension of Python?",
        "options": ["A) .pt", "B) .py", "C) .python", "D) .p"],
        "answer": "B"
    },
    {
        "question": "2. Which data type stores multiple values?",
        "options": ["A) int", "B) float", "C) list", "D) bool"],
        "answer": "C"
    },
    {
        "question": "3. Which symbol is used for comments in Python?",
        "options": ["A) //", "B) #", "C) <!--", "D) **"],
        "answer": "B"
    },
    {
        "question": "4. What does len() function do?",
        "options": ["A) Add values", "B) Delete value", "C) Count length", "D) Sort list"],
        "answer": "C"
    },
    {
        "question": "5. Which keyword is used for loop?",
        "options": ["A) loop", "B) repeat", "C) for", "D) iterate"],
        "answer": "C"
    }
]

score = 0

print("📘 Welcome to Python Quiz App\n")

# Loop through each question
for q in quiz_questions:
    print(q["question"])
    
    for option in q["options"]:
        print(option)
    
    user_answer = input("Enter your answer (A/B/C/D): ").upper()
    
    if user_answer == q["answer"]:
        print("✅ Correct!\n")
        score += 1
    else:
        print("❌ Wrong! Correct answer is:", q["answer"], "\n")

# Final Result
print("🎯 Quiz Completed!")
print("Your Final Score:", score, "/", len(quiz_questions))

percentage = (score / len(quiz_questions)) * 100
print("Percentage:", percentage, "%")

if percentage >= 70:
    print("Result: Pass 🎉")
else:
    print("Result: Fail ❌")
````
