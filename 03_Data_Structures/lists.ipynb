## 📌 1. List

```python
# A list is an ordered, mutable collection of items that can store multiple values of any data type in a single variable.
# Syntax:
# Creating a list
# list_name = [item1, item2, item3, ...]

# Creating a List
fruits     = ["Apple", "Banana", "Mango", "Orange"] # List of Strings
marks      = [50, 20, 40, 40, 80]              # List of Integers
mixed      = ["Ali", 25, 3.5, True]            # list can store mixed data types or List with Different Data Types
empty_list = []                                 # empty list
```

```python
#Define List
fruits = ["Apple", "Banana", "Mango", "Orange"]
print(fruits)
```

```python
print(fruits[0]) #access first element of the list using index number
```

## Functions(Methods) of List

## 1. append()

```python
fruits.append("Kiwi") #add element at the end of the list
print(fruits)
```

## 2. insert()

```python
fruits.insert(1,"Cherry") #add element at first index of the list
print(fruits)
```

## 3. remove()

```python
fruits.remove("Banana") #remove banana from the list using remove method
print(fruits)
```

## 4. extend()

```python
fruits.extend(["Watermelon","Pomegranate","Strawberry"]) #add mulitple elements at the end of the list
print(fruits)
```

## 5. pop()

```python
fruits.pop() #remove last element from the list
```

```python
print(fruits)
```

```python
fruits.pop(1) #remove element from index 1
```

```python
print(fruits)
```

## 6. reverse()

```python
#Elemnts in reverse order
fruits.reverse()
print(fruits)
```

## 7. sort()

```python
#Sort/Arrange Element
fruits.sort() #arrange element in ascending order
print(fruits)
```

```python
fruits.sort(reverse = True) #arrange elements in descending order
print(fruits)
```

## 8. len()

```python
#Count the Number of Elements in the list
length = len(fruits)
print(length)
```

## 9. index()

```python
#Find position of value
print(fruits.index("Apple")) #apple is at index 5 in the above updated list
```

## 10. count()

```python
#Counts how many times a value appears in the list.
print(fruits.count("Apple"))
```

## 11. copy()

```python
#Makes a duplicate list
new_fruits = fruits.copy()
print(new_fruits)
```

## 12. clear()

```python
#Removes all elements from the list(duplicate list) and make empty list
new_fruits.clear()
print(new_fruits)
```

## Slicing (getting a part (portion) of a list

```python
#print the original list of fruits
print(fruits)
```

```python
#Slicing
fruits = fruits[1:4] #slice (start from index 2(orange) to index 4(Watermelon) ) but last index is not included
print(fruits)
```

```python
fruits[1] = "Grape"  #replace element at index 1(orange) with Grape
print(fruits)
```

```python
#delete element from the fruits list using index number
del fruits[0]
print(fruits)
```

```python
#delete permanently list instead of making empty list 
del new_fruits
```

## Create an empty list to store marks

```python
marks = []
# Loop 5 times to take marks for 5 subjects
for i in range(5):
    n = int(input(f"Enter Marks for subject {i+1} = "))
    marks.append(n)  # Add the entered marks into the list
total_marks = sum(marks)
avg = total_marks/len(marks)
lowest_marks = min(marks)
highest_marks = max(marks)
print("Total Marks = ", total_marks)
print("Average = ", avg)
print("Minimum Marks = ", lowest_marks)
print("Maximum Marks = ", highest_marks)
```
