## 📌 2. Tuple

```python
# A tuple is an ordered, immutable collection of items that can store multiple values of any data type in a single variable.
# Immutable means tuple elements cannot be changed after creation.

# Syntax:
# tuple_name = (item1, item2, item3, ...)

# Creating a Tuple
fruits      = ("Apple", "Banana", "Mango", "Orange")
marks       = (50, 20, 40, 40, 80)
mixed       = ("Ali", 25, 3.5, True)    # mixed data types
empty_tuple = ()                         # empty tuple
single      = ("Apple",)                 # single item — comma is required
```

```python
# Define Tuple
fruits = ("Apple", "Banana", "Mango", "Orange")
print(fruits)
```

```python
# Access Tuple Elements (Indexing)
print(fruits[0])  # Access first element of the tuple using index number
print(fruits[2])  # Access third element
```

## Tuple Functions (Built-in Functions)

## 1. len()

```python
# Count the number of elements in the tuple
length = len(fruits)
print(length)
```

## 2. index()

```python
# Find the position of a value
print(fruits.index("Banana"))
```

## 3. count()

```python
# Count how many times a value appears
marks = (50, 20, 40, 40, 80)
print(marks.count(40))
```

## Tuple Slicing (Getting a Portion of a Tuple)

```python
marks = (50, 20, 40, 40, 80)
print(marks[1:4])  # Start from index 1 to 3
```

## Loop Through Tuple

```python
# Tuple containing days of the week
days = ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday")
for day in days:  # loop through each day
    # Check if the day is weekend
    if(day == "Saturday" or day == "Sunday"):
        print(day, "... Holiday")   # print holiday message
    else:
        print(day)                # print normal day
```

## Update Tuples

```python
# Tuples are immutable, which means their elements cannot be changed, added, or removed after creation.
# However, there are some workarounds to update tuple values.
```

## 1. Direct Update (Not Allowed)

```python
marks = (50, 20, 40, 40, 80)
# marks[1] = 70   # ❌ This will cause an error because tuples are immutable.
```

## 2. Convert Tuple to List → Update → Convert Back to Tuple

```python
# Original tuple containing marks
marks = (50, 20, 40, 40, 80)
marks = list(marks) # Convert the tuple into a list because tuples are immutable (cannot be changed)

marks[1] = 70 # Update the value at index 1 (second element)
marks.append(95) # Add a new element at the end of the list
marks = tuple(marks) # Convert the list back into a tuple

# Print the updated tuple
print(marks)
```
