## Conditional Statements

```python
"""Conditional statements are used to execute a block of code based on a condition. 
If the condition is True, the block of code is executed, otherwise the block of code is skipped."""
```

## Types of Conditional Statements

```python
# Conditional Statements
# │
# ├── 1. if
# ├── 2. if / else
# ├── 3. if / elif / else
# ├── 4. Nested if
```

## 1. if Statement

```python
# The if statement executes a block of code only if the given condition is True.

# Syntax
# if(condition):
#     statements

#Check Pass Status
marks = int(input("Enter Your Marks = "))
if(marks > 40): # Check if marks are greater than 40
    print("PASS")
#Driving License Eligibility
age = int(input("Enter Your Age = "))
if (age >= 18):  # Check if age is 18 or above
    print("You are Eligible for Driving License.")
```

## 2. if-else Statement

```python
# The if-else statement executes the if block when the condition is True, otherwise the else block is executed.

# Syntax
# if(condition):
#     statements 
# else:
#     statements

# Example 1: Check Pass or Fail
marks = int(input("Enter Your Marks = "))
if(marks > 40): # Check if marks are greater than 40
    print("PASS") # Output: Pass if condition is true
else:
    print("FAIL") #output: fail if condition is false.


# Example 2: Driving License Eligibility
age = int(input("Enter Your Age = "))
if (age >= 18):  # Check if age is 18 or above
    print("You are Eligible for Driving License.")
else:
    print("You are not Eligible for Driving License.")

# Example 3: Even or Odd Number
num = int(input("Enter a Number = "))
if(num%2 == 0):  #if remainder is zero --> even number
    print("Number is Even")
else:
    print("Number is Odd") #if remainder is not equal to zero --> odd number
```

## 3. if-elif-else Statement

```python
#The if-elif-else statement is used when there are more than two conditions to check.
# Syntax:
# if condition1:
#     statements
# elif condition2:
#     statements
# elif condition3:
#     statements
#         .
#         .
#         .
# else:
#     statements

# Program to check if a number is Positive, Negative, or Zero
num = int(input("Enter a Number = "))
if(num > 0):
    print("Positive Number")
elif(num < 0):
    print("Negative Number")
else:
    print("Equal to Zero")


# Program that takes three numbers from the user and prints the largest number among them.

num1 = int(input("Enter First Number = "))
num2 = int(input("Enter Second Number = "))
num3 = int(input("Enter Third Number = "))

if(num1 > num2 and num1 > num3): # Check if num1 is greater than both num2 and num3
    print(f"{num1} is the Largest Number")
    
elif(num2 > num1 and num2 > num3): # Check if num2 is greater than both num1 and num3
    print(f"{num2} is the Largest Number")
    
else: # If neither num1 nor num2 is the largest, num3 must be the largest
    print(f"{num3} is the Largest Number")
```

## 4. Nested if Statement

```python
# An if statement inside another if statement.
# Syntax:
# if condition1:
#     if condition2:
#         # runs only if both condition1 and condition2 are True
#     else:
#         # runs if condition1 True but condition2 False
# else:
#     # runs if condition1 is False

# Example 1 — Simple Login System
username = "umar"
password = "umar1234"

if username == "umar":
    if password == "umar1234":
        print("Login successful")
    else:
        print("Wrong password")
else:
    print("Username not found")

# Example 2 — Check Positive/Negative and Even/Odd Number
number = int(input("Enter any Number = "))
if (number > 0):
    if (number % 2) == 0:
        print("Positive Even number")
    else:
        print("Positive Odd number")
else:
    print("Negative number")
```
