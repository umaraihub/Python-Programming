## Loops

```python
'''A loop is a control structure that repeatedly executes a block of statements while a given condition is true or 
until a sequence ends.'''
```

## 📌 Types of Loops in Python

```python
# Loops in Python
# │
# ├── 1. for Loop        → repeats a fixed number of times
# ├── 2. while Loop      → repeats until condition is False
# └── 3. Loop Control
#          ├── break     → exit the loop immediately
#          ├── continue  → skip current iteration
#          └── pass      → do nothing, placeholder
```

## 1. for Loop

```python
# It is used when the number of iterations is known in advance. OR You know exactly how many times to repeat
# for variable in sequence:
#     statements
```

##  Example 1 — Print Each Character of a String

```python

# Loop through each character in a string

name = "Umar"

for letter in name:
    print(letter) #print each character in new line

for letter in name:
    print(letter,end=" ") #print each character with space in single line
    
print() #for line break
```

## Example 2 — Print Numbers and Asterisk Patterns Using range()

```python
# range(stop)           # 0 to stop-1
# range(start, stop)    # start to stop-1
# range(start, stop, step)  # start to stop-1 with ste
for i in range(5): #start=0 stop = 5(exclude last value 5) increment 1 by default
    print(i)

for i in range(1,5): 
    print(i*"*")
```

```python
for i in range(5,0,-1): # Loop from 5 down to 1 (greater than 0 or exclude 0), decreasing by 1 each time
    print(i)

for i in range(5,0,-1):
    print(i*"*")
```

```python
n = int(input("Enter Number of Rows = "))  # number of rows in the top half

# Loop through each row
for i in range(1, 2*n):
    if i <= n:
        # Top half
        for j in range(n - i):
            print(" ", end="")  # print spaces
        for j in range(i):
            print("* ", end="")  # print stars with space
    else:
        # Bottom half
        for j in range(i - n):
            print(" ", end="")  # print spaces
        for j in range(2*n - i):
            print("* ", end="")  # print stars with space
    print()  # new line (print() at the end of outer loop)
```

```python
n = int(input("Enter number of rows: "))

for i in range(n, 0, -1):
    # Print spaces
    for j in range(n - i):
        print(" ", end="")
    # Print stars
    for j in range(i):
        print("*", end="")
    print()  # move to next line
```

## 2. While Loop

```python
# It is used when the number of iterations is not known in advance.
# Syntax:
# while condition:
#     statements
```

## Example 1 — Print Numbers and Asterisk Patterns

```python
i = 1 #Initialization
while (i < 6): #Test Condition
    print(i) #statements
    i = i + 1  # increment i to avoid infinite loop (increment/decrement)
```

## Example 3 — User Input Validation

```python
password = input("Enter Password")      

while password != "1234":              # keep asking until correct password
    password = input("You entered Incorrect Password.\nPlease Enter Correct Password = ")  # take input from user
print("Access Granted ✅")             # prints when correct password entered
```

## 3. Loop Control Statements

```python
# Loop control statements are used to change the normal flow of a loop. 
# Python has three loop control statements:

# 1. break    — exits the loop immediately
# 2. continue — skips the current iteration and moves to next
# 3. pass     — does nothing, used as a placeholder
```

## 3a. break Statement

```python
for i in range(6):
    if i == 4:
        break
    print(i)
```

## 3b. continue Statement

```python
for i in range(6):
    if i == 4:
        continue    # skip number 4
    print(i)
print("============")
```

## 3c. pass Statement

```python
# Write a Python program using a for loop from 1 to 10 that uses pass for even numbers and prints only odd numbers.
for i in range(1,5):
    if(i%2 == 0):
        pass # Do nothing for even numbers
    else:    
        print(i)
```

# Program to demonstrate pass, continue, and break in a loop

```python
for i in range(1, 10):
    # If number is even, use pass (do nothing)
    if i % 2 == 0 and i != 8:
        print(i, "is even - pass executed (nothing happens)")
        pass  # placeholder, loop continues normally

    # If number is divisible by 3, skip this iteration
    elif i % 3 == 0:
        print(i, "is divisible by 3 - continue executed (skipped)")
        continue  # skip the rest of the loop for this iteration

    # If number is 8, exit the loop completely
    elif i == 8:
        print(i, "found - break executed (loop stops)")
        break  # exit the loop immediately

    # For all other numbers, just print normally
    else:
        print(i, "is an odd number")
```
