## String Operations

```python
first_name = "Muhammad"
last_name  = "Umar"

# Concatenation
print(first_name + " " + last_name)    # Muhammad Umar

# Repetition
print("Ha" * 3)               # HaHaHa

# Length
print(len(first_name))             # 8

# Indexing — access single character (starts from 0)
print(first_name[0])               # M
print(first_name[-1])              # d  
# -1 gets the last character of the string; -2 would get the second last character, -3 the third last, and so on (negative indexing counts from the end)

# Slicing — access part of string
# Prints characters from index 0 to 4 (start index included, end index excluded)
print(first_name[0:5])             # Muham

# Prints characters from index 2 to the end of the string
print(first_name[2:])              # hammad

# Prints characters from the beginning (index 0) up to index 3
print(first_name[:4])              # Muha
```

## String Methods

```python
course = "  python programming  "   # String with extra spaces at the beginning and end
print(course.upper())               # Converts all characters in the string to uppercase
print(course.lower())               # Converts all characters in the string to lowercase
print(course.title())               # Converts the first letter of each word to uppercase
print(course.replace('python','c++'))  # Replaces the word 'python' with 'c++' in the string
print(course.rstrip())              # Removes spaces from the right (end) side of the string
print(course.lstrip())              # Removes spaces from the left (start) side of the string
```

```python
course = course.strip()             # Removes spaces from both the beginning and end and stores the cleaned string back into variable
print(course)                       # Prints the string after removing spaces from both sides
print(course.split(" "))            # Splits the string into a list of words using space as separator
print(course.count("p"))            # Counts how many times the letter 'p' appears in the string
print(course.startswith("python"))  # Checks if the string starts with the word 'python' (returns True or False)
print(course.endswith("python"))    # Checks if the string ends with the word 'python'
print(course.endswith("programming")) # Checks if the string ends with the word 'programming'
```

## Type Casting

```python
#Type casting is the process of converting a value from one data type to another data type in a program.

# String to int
age = "25"  # age is a string
age = int(age)  # converting string to integer
print(age + 5)       # 30
print(type(age))

# Int to string
score = 88
score = str(score)
print("Score: " + score)   # Score: 88
print(type(score))

# Int to float
price = 150
x = float(price)
print(x)             # 150.0
print(type(x))

# Float to int — truncates (removes decimal)
pi = 3.14
y = int(pi)
print(y)             # 3
print(type(y))
```

## Operators

```python
#An operator is a symbol that performs an operation on one or more values. The values it works on are called operands.

#result = 10 + 5
#        ──  ─  ─
#        │   │  └── operand 2
#        │   └───── operator
#        └───────── operand 1


## Python Operator Types
#Python Operators
# │
# ├── 1. Arithmetic Operators       + - * / // % **
# ├── 2. Assignment Operators       = += -= *= /= //= %= **=
# ├── 3. Comparison Operators       == != > < >= <=
# ├── 4. Logical Operators          and  or  not
# ├── 5. Identity Operators         is   is not
# ├── 6. Membership Operators       in   not in
# └── 7. Bitwise Operators          & | ^ ~ << >>
```

## 1. Arithmetic Operators

```python
#Arithmetic Operators are used to perform arithmetic operations.
"""Arithmetic Operators are as follows: 
+,-,*,/,%,//,**"""
```

```python
num1 = int(input("Enter First Number = "))
num2 = int(input("Enter Second Number = "))
add = num1 + num2
diff = num1 - num2
prod = num1 * num2
div = num1 / num2
rem = num1 % num2
fdiv = num1 // num2
exp = num1 ** 2
print("SUM = ", add)
print("Difference = ", diff)
print("Product = ", prod)
print("Division = ", div)
print("Remainder = ", rem)
print("Floor Division = ", fdiv)
print("Exponent = ", exp)
```

## 2. Comparison/ Relational Operators

```python
#Comparison operators are used to compare the values of two variables
#Comparison Operators are : >, <, >=, <=, !=, ==
num1 = int(input("Enter First Number = "))
num2 = int(input("Enter Second Number = "))
print(num1 > num2)
print(num1 < num2)
print(num1 >= num2)
print(num1 <= num2)
print(num1 != num2)
print(num1 == num2)
```

## 3. Logical Operators

```python
#Logical operators are used to combine multiple conditions or invert a condition.
#The result is always a Boolean value: True or False.
#Logical Operators are: and, or, not


num = int(input("Enter Any Number = "))

# Check if the number is greater than 0 AND less than 10
# This will print True only if both conditions are True
print(num > 0 and num < 10)

# Check if the number is greater than 0 OR less than 10
# This will print True if at least one of the conditions is True
print(num > 0 or num < 10)

# Apply NOT to the condition (num > 0 and num < 10)
# This will print the opposite of the first condition's result
print(not(num > 0 and num < 10))
```

## 4. Membership Operators

```python
#Membership operators are used to check if a value exists inside a sequence (string, list, tuple, etc.)
#Membership Operators are : in, not in
name = "Muhammad Umar"
print("Umar" in name)        # True
print("Ali" in name)         # False
print("Ali" not in name)     # True
print("m" in name)   # True
```

## 5. Bitwise Operators

```python
#Bitwise operators are used to perform operations on the binary bits.
#Bitwise Operators are: & (AND), | (OR), ^ (XOR), ~ (NOT), << (Left Shift), >> (Right Shift)

#     2⁷, 2⁶, 2⁵, 2⁴, 2³, 2², 2¹, 2⁰
#25:  0   0   0   1   1   0   0   1 = 24 + 1 => 25
#65:  0   1   0   0   0   0   0   1 = 64 + 1 => 65
a = 25 # binary: 00011001
b = 65 # binary: 01000001

#     2⁷, 2⁶, 2⁵, 2⁴, 2³, 2², 2¹, 2⁰
#25:  0   0   0   1   1   0   0   1 = 24 + 1 => 25
#65:  0   1   0   0   0   0   0   1 = 64 + 1 => 65

'''_________________________________________________
a&b=  0   0   0   0   0   0   0   1 = 1          # Bitwise AND (&) - sets bit to 1 only if both bits are 1
_________________________________________________
a\b=  0   1   0   1   1   0   0   1 = 64+16+8+1 => 89  # Bitwise OR (|) - sets bit to 1 if at least one bit is 1
_________________________________________________
a^b=  0   1   0   1   1   0   0   0 = 64+16+8 => 88  # Bitwise XOR (^) - sets bit to 1 if bits are different
_________________________________________________'''


print("Bitwise AND Operator = ", a & b)
print("Bitwise OR Operator = ", a | b)
print("Bitwise XOR Operator = ", a ^ b)


# ~a = -(a+1) = -26
print("Bitwise Not Operator = ", ~a)  # Output: -26 # Bitwise NOT (~) - inverts all bits (two's complement, so negative)

# Left Shift (<<) - shifts bits to the left, fills 0 on right
# a << 1: 00011001 << 1 -> 00110010 = 50

#HINT: IF a<<1 means one shift then multiply number with 2.
#multiply by 2 for each shift. (25*2 = 50)
print("Left Shift Operator = ", a << 1)  # Output: 50 


# Right Shift (>>) - shifts bits to the right, fills 0 on left
# b >> 1: 01000001 >> 1 -> 00100000 = 32

# divide by 2 for each shift. (64/2 = 32)
print("Right Shift Operator = ", b >> 1)  # Output: 32
```

## Operator Precedence (BODMAS for Python)

```python
# ( )          ← Brackets always first
# **           ← Powers
# * / // %     ← Multiply & Divide
# + -          ← Add & Subtract
# == != > < >= <=  ← Compare
# not          ← Logical NOT
# and          ← Logical AND
# or           ← Logical OR (always last)
print(10 - 2 ** 3)     # 10 - 8 = 2  — exponent first (2³ = 8)
print(10 / 2 + 3 * 4)  # division and multiplication left to right → 5.0 + 12 = 17.0
print(100 - 50 / 2 + 5 * 3) #100-25.0+15 =>75.0 + 15 => 90.0
```
